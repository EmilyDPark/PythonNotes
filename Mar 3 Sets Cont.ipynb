### Review
#### Consider Problem 4 on page 48 of the finite mathematics text. Write the code which generates all 256 possible strings which can be made using the letters of “STOP”. Place them in a list named “words”. Print the list and its length.

```python
words = []
for l1 in "STOP":
    for l2 in "STOP":
        for l3 in "STOP":
            for l4 in "STOP":
                word = l1 + l2 + l3 + l4
                words.append(word)
print(words)
```

```python
print(len(words))
```

### Avoid Repetition 1
#### The author’s hint indicated that there were only 24 strings. This follows from a constraint we did not impose in the code above. The constraint is that a letter can occur only once in a string.

#### There are at least two ways we could modify the code above to impose this constraint.

#### One method is to check each string generated by the crude algorithm and see if it satisfies the constraint before appending it to the list. Write code to impose the constraint this way and call the resulting list “words1”. Print the list and its length.

```python
words1 = list()

stop = set(['S', 'T', 'O', 'P'])
wordSet = set()

for word in words:
    for letter in word:
        wordSet.add(letter)
    if wordSet & stop == stop:
        words1.append(word)
    wordSet.clear()
    
print(words1)
```

```python
print(len(words1))
```

```python
words1 = list()

stop = set(['S', 'T', 'O', 'P'])

for word in words:
    if set(list(word)) & stop == stop:
        words1.append(word)
    
print(words1)
```

### Avoid Repetition 2
#### An alternative method is to reduce the set of letters available to successive levels of the nested for loops. Write code which uses this method and use it to produce a list named words2. Print this list and its length.

```python
# Professor's Answer

words2 = []
wset = set(list("STOP"))
for l1 in wset:
    for l2 in wset - {l1}:
        for l3 in wset - {l1, l2}:
            for l4 in wset - {l1, l2, l3}:
                word = l1 + l2 + l3 + l4
                words2.append(word)
print(words2)
```

### Comparison
#### Both of these methods seem to have worked. Are the two lists really equivalent? Write code to answer this question.

```python
set(words1) == set(words2)
```

```python
print(set(words1) & set(words2))
```

The remaining items in the tuple are one-character strings giving the winners of the primaries in the four states. You should use the same pattern of nested for loops inherent in the table as it is presented in the text.

```python
# Professor's Answer

possibilities = []
i = 0
for NH in "AC":
    for MN in "ABC":
        for WI in "AB":
            for CA in "ABC":
                i += 1
                outcome = (i,NH,MN,WI,CA)
                possibilities.append(outcome)
for i in possibilities:
    print(i)
```

### Exercise

### Exercise
#### Write a function counter(cand,outcome) which returns the count of a candidate’s win in a particular outcome. For example counter(“A”, (1,“A”,“A”,“A”,“A”)) should return 4.

#### Test your function with a few of the possibilities you see above.

```python
def counter(cand, outcome):
    count = 0;
    for winner in outcome:
        if cand == winner:
            count += 1
    return count
```

```python
print(counter('A', (1, 'A', 'A', 'A', 'A')))
```

### Example 1
#### Use set comprehension to produce the set described in Example 1. Display it in numerical order.

```python
# Professor's Answer

Example1 = {OC for OC in Universe if 
(counter("B",OC) > counter("A",OC)) and 
(counter("B",OC) > counter("C",OC))}

for OC in sorted(Example1):
    print(OC)
```
